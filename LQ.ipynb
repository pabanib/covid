# Comportamientos del coeficiente de localización

```python
import numpy as np
import matplotlib.pyplot as plt 
```

### Comportamientos de LQ ante cambios de $x_i$

```python
x = 280
t = 1320
ti = 120

xi = np.linspace(0,120)
lq = []
def lq_(xi, ti = ti, t = t, x = x ):
    return (xi/ti)/(x/t)
for i in xi:
    lq.append(lq_(i))
plt.plot(xi,lq)
#plt.axvline(p,0,1, color = 'r')
```

Ante variaciónes de la cantidad de afectados el coeficiente de localización se comporta de manera lineal

```python
ni = np.linspace(20,1320)
lq = []
for i in ni:
    lq.append(lq_(20,i))
plt.plot(ni,lq)
    
```

Ante variaciónes de la población decrece rapidamente

## Comportamiento de la varianza del coeficiente de localizaciön

```python
pi = 0.5
p = 0.4
ni = 30
n = 300

def var(pi):
    var = (pi*(1-pi)/(ni*p**2))+((pi**2)*(1-p)/(n*p**3))+(2*(pi**2)*(1-pi)/(n*p**3))
    return var
```

```python
espacio = np.linspace(0,1)
y = []
for i in espacio:
    y.append(var(i))
```

```python
plt.plot(espacio,y)
plt.axvline(0.4,0,1, color = 'r')
```

Manteniendo la cantidad poblacional de la unidad fija $n_i$ la varianza del LQ tiende a incrementarse en valores menores de $r_i < r$ e incluso el máximo lo logra pasando a este para luego en valores mucho mas grandes que el pobalcional empieza a disminuir

### Probando con otra $r$ poblacional

```python
pi = 0.5
p = 0.1
ni = 30
n = 300

espacio = np.linspace(0,1)
y = []
for i in espacio:
    y.append(var(i))
plt.plot(espacio,y)
plt.axvline(p,0,1, color = 'r')
```

Ante valores pequeños de $p$ la varianza se incrementa en casi todos los puntos, a medida que crece $p$ se vuelve mas parábolica

### Variaciones de la población de la unidad $n_i$

```python
pi = 0.5
p = 0.4
ni = 30
n = 300

def var(pi, ni):
    x = 20
    pi = x/ni
    var = (pi*(1-pi)/(ni*p**2))+((pi**2)*(1-p)/(n*p**3))+(2*(pi**2)*(1-pi)/(n*p**3))
    return var

espacio = np.linspace(20,300)
y = []
for i in espacio:
    y.append(var(0.2,i))

plt.plot(espacio,y)
#plt.axvline(p,0,1, color = 'r')
```

A medida que crece $n_i$ manteniendo fijo el valor de $x_i$ la varianza crece rápido en los primeros tramos y luego decrece muy rapidamente
